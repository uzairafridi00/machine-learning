# Seaborn and plotting

```python
# import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Important things for plotting according to grammar of graphics

- Data
- Type of plot
- X Axis
- Y Axis
- Grouping (hue/color) based on categorical variable
- Size of Plot
- Color pallets of the plot
- Style of the plot
- How do you interpret?

```python
# load the data

df = sns.load_dataset('titanic')
df.head()
```

```python
df.info()
```

- Scatter Plot

```python
sns.scatterplot(data=df, x='age', y='fare', hue='sex')
```

- Size of Figure

```python
# plot figure size
plt.figure(figsize=(16,9))

sns.scatterplot(data=df, x='age', y='fare', hue='sex')
```

- Change the Colors

```python
# Change the scatter point colors
palette = {'male': 'green', 'female': 'orange'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=palette)

# add labels
plt.xlabel('Age')
plt.ylabel('Fare')

# add title
plt.title('Scatter Plot of Age vs Fare')
```

- Saving the Plot

```python
# Change the scatter point colors
palette = {'male': 'green', 'female': 'orange'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=palette)

# add labels
plt.xlabel('Age')
plt.ylabel('Fare')

# add title
plt.title('Scatter Plot of Age vs Fare')

# Save a plot 300dpi, 600dpi, 150dpi png
plt.savefig('./outputs/scatter_plot_age_vs_fare.png', dpi=300)

# Save a plot in Vector File such as SVG
plt.savefig('./outputs/scatter_plot.svg')
```

- Style of a Plot

```python
# style of plot
sns.set_theme(style='darkgrid', font_scale=1, font='Times New Roman')

# Change the scatter point colors
palette = {'male': 'green', 'female': 'orange'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=palette)

# add labels
plt.xlabel('Age')
plt.ylabel('Fare')

# add title
plt.title('Scatter Plot of Age vs Fare')
```

## Interpretation

1. We have age of passengers at X-Axis and Fare at Y-Axis.
2. The scatter plot shows the relationship between age and fare of passengers.
3. We further grouped the passenger by their genders such as Male (green points) and Female (orange points).
4. Two Male passengers gave the highest fare, and they might be travelling in first class.
5. The oldest person at titanic is also a Male having 80 years of age as shown in the scatter plot (fig 1).

# Box Plot

```python
df['class'].unique()
```

```python
# style of plot
sns.set_theme(style='darkgrid', font_scale=0.8, font='Times New Roman')

# Change the scatter point colors
palette = {'First': 'green', 'Second': 'orange', 'Third': 'blue'}
sns.boxplot(data=df, x='sex', y='age', hue='class', palette=palette)

# add labels
plt.xlabel('Sex')
plt.ylabel('Age')

# add title
plt.title('Scatter Plot of Age vs Fare')

# change the legend position and title
plt.legend(loc='upper right', title='Class Name')
```

- Grow with Size

```python
# plot figure size
plt.figure(figsize=(16,9))

# style of plot
sns.set_theme(style='darkgrid', font_scale=1, font='Times New Roman')

# Change the scatter point colors
palette = {'male': 'green', 'female': 'orange'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', size='age', palette=palette)

# add labels
plt.xlabel('Age')
plt.ylabel('Fare')

# add title
plt.title('Scatter Plot of Age vs Fare')
```

- Boxen Plot

```python
# plot figure size
plt.figure(figsize=(16,9))

# style of plot
sns.set_theme(style='darkgrid', font_scale=0.8, font='Times New Roman')

# Change the scatter point colors
palette = {'First': 'green', 'Second': 'orange', 'Third': 'blue'}
sns.boxenplot(data=df, x='sex', y='age', hue='class', palette=palette)

# add labels
plt.xlabel('Gender')
plt.ylabel('Age')

# add title
plt.title('Scatter Plot of Age, based on Gender and Class Name of Passengers')

# change the legend position and title
plt.legend(loc='upper right', title='Class Name')

# change the x tick labels
plt.xticks(ticks=[0, 1] ,labels=["Male", "Female"], rotation=45)
```

# Plotly

```python
# make scatter plot

import plotly.express as px

fig = px.scatter(df, x='age', y='fare', color='sex')
fig.show()
```

- Bar Plot 

```python
# plot figure size
plt.figure(figsize=(16,9))

# style of plot
sns.set_theme(style='darkgrid', font_scale=1.5, font='Calibri')

# Change the scatter point colors
palette = {'First': 'green', 'Second': 'orange', 'Third': 'blue'}
sns.barplot(data=df, x='sex', y='age', hue='class', palette=palette, capsize=0.1)

# add labels
plt.xlabel('Gender')
plt.ylabel('Age')

# add title
plt.title('Bar Plot of Age, based on Gender and Class Name of Passengers')

# change the legend position and title
plt.legend(loc='upper right', title='Class Name')

# change the x tick labels
plt.xticks(ticks=[0, 1] ,labels=["Male", "Female"], rotation=45)
```

```python
# bar plot using seaborn

# style of a plot
sns.set_theme(style='darkgrid', font_scale=1, font='Times New Roman')
# scatter plot size of figure size
plt.figure(figsize=(12,6)) # a4 page size 16:9

# change the scatter point colors
palette = {'First': 'green', 'Second': 'red', 'Third': 'blue'}
sns.violinplot(data=df, x='sex', y='age', hue='class', palette=palette)
# labels
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Bar plot of Age, based on Gender and Class Name of Passengers')

# change legend position and title
plt.legend(loc='upper right', title='Class Name')

# change the x-ticks names
plt.xticks(ticks=[0, 1], labels=["Male", "Female"], rotation=0)
```
