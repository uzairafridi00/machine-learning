## Imports

```python
import pickle
import joblib
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

## Read the Data

```python
df = pd.read_csv('/content/housing.csv').iloc[:, :-1].dropna()
df
```

```python
X = df.drop(columns='median_house_value')
y = df.median_house_value.copy()

print(X.shape, y.shape)
```

## Train the Model

```python
model = LinearRegression()
model.fit(X, y)
```

```python
y_pred = model.predict(X)
y_pred
```

```python
r2_score(y, y_pred)
```

```python
model
```

## Pickle

```python
with open('model.pkl', 'wb') as f:
  pickle.dump(model, f)
```

```python
with open('model.pkl', 'rb') as f:
  pickle_model = pickle.load(f)
```

```python
pickle_model
```

```python
y_pred_pickle = pickle_model.predict(X)
y_pred_pickle
```

```python
r2_score(y, y_pred_pickle)
```

## Joblib

```python
joblib.dump(model, 'model.joblib')
```

```python
joblib_model = joblib.load('model.joblib')
joblib_model
```

```python
y_pred_joblib = joblib_model.predict(X)
y_pred_joblib
```

```python
r2_score(y, y_pred_joblib)
```
