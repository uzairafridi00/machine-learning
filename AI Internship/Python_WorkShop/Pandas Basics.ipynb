### Grouping

Grouping works similar to "group by" as in databases. Pandas also provides a "group by" function which serves the same purpose.

```python
# Set display options
pd.set_option('display.float_format', '{:.6f}'.format)  # To display floats without scientific notation
pd.set_option('display.max_columns', None)  # To display all columns
```

```python
# Remove any non-numeric characters and convert to float
df_cars['Ex-Showroom_Price'] = df_cars['Ex-Showroom_Price'].replace({'Rs.': '', ',': ''}, regex=True).astype(float)
```

```python
df_cars.head(2)
```

```python
# getting average prices of cars on Drivetrain
df_cars.groupby('Drivetrain')['Ex-Showroom_Price'].mean()
```

```python
df_cars.groupby('Variant')['Ex-Showroom_Price'].mean()
```

```python
df_cars.groupby(['Make', 'Drivetrain'])['Ex-Showroom_Price'].mean()
```

```python
# shows in dataframe form
group = df_cars.groupby(['Make', 'Variant', 'Drivetrain'])['Ex-Showroom_Price'].mean().reset_index()
```

```python
group.sort_values(['Ex-Showroom_Price'], ascending=False)
```

```python
df_cars.dtypes
```

### Named Aggregation

```python
# creates new column with avg_price name and the value will be stored there of price mean
df_cars.groupby(['Make']).agg(avg_price = ('Ex-Showroom_Price', 'mean'),
                                            max_price = ('Ex-Showroom_Price', 'max'),
                                            min_price = ('Ex-Showroom_Price', 'min')).reset_index()
```

## Pivoting and Melting

Pivoting is the process of transforming or reshaping data from a long format to a wide format. It allows you to create a new DataFrame where one or more columns become the new index, and another column's values are spread across multiple columns.


Melting is the reverse operation of pivoting. It transforms data from a wide format to a long format, allowing you to consolidate multiple columns into key-value pairs.

- Pivoting
- Long format (Vertically) convert to Wide format (Horizontally)

```python
#group
```

```python
pivoted = pd.pivot_table(group, index='Make', columns='Variant', values='Ex-Showroom_Price', aggfunc='mean')
```
