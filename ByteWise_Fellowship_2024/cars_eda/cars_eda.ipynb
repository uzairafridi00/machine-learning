## **EDA on Cars Dataset**

```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# set color theme
sns.set_theme(color_codes=True)
```

```python
# load the data
df = pd.read_csv('cars_data.csv')

# display the data
df.head(2)
```

```python
# display the bottom 2 rows
df.tail(2)
```

```python
# check the types of data
df.dtypes
```

```python
# dropping irrelevant columns
df = df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
```

```python
# display the data
df.head(3)
```

```python
# renaming the columns
df = df.rename(columns={"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission",
                        "Driven_Wheels": "Drive Mode","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price" })
```

```python
df.head(5)
```

```python
# Dropping the duplicates rows
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print('Number of Duplicated Rows: ', duplicate_rows_df.shape)
```

```python
 # Used to count the number of rows
df.count()
```

```python
df = df.drop_duplicates()
df.head(5)
```

```python
df.count()
```

```python
# Dropping the missing or null values
print(df.isnull().sum())
```

```python
df = df.dropna()
df.count()
```

```python
# After dropping the values
print(df.isnull().sum())   
```

```python
# Detecting Outliers
sns.boxplot(x=df['Price'])
```

```python
sns.boxplot(x=df['HP'])
```

```python
sns.boxplot(x=df['Cylinders'])
```

```python
# Remove using IQR
# Assuming df is your DataFrame
numeric_cols = df.select_dtypes(include=[np.number]).columns
Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df = df[numeric_cols][~((df[numeric_cols] < (Q1 - 1.5 * IQR)) |(df[numeric_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

```python
# Plot different features against one another (scatter), against frequency (histogram)

df['Cylinders'].value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make')
```

```python
plt.figure(figsize=(10,5))
c= df.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['HP'], df['Price'])
ax.set_xlabel('HP')
ax.set_ylabel('Price')
plt.show()
```
