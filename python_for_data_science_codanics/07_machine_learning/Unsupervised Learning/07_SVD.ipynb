# **SVD (Singular value decomposition)**

1. Use for Dimesnality Reduction.
2. Unsupervide ML
3. Use for Data compression.
4. Reducde the Noise
5. Also called filtering technique for features

It's a linear algebric method, use to decompose the matrix into 3 matrices. In these 2 are orthogonal matrix.

### Matrix Factorization

1. Row and Columns (5x6)
2. Convert high dimension matrix into lower dimensions. It's called decomposition of matrices.
3. Then these small matrix are used in Linear equation.
4. 2 Square matrix is called orthogonal matrix and their combination makes the real matrix.

- A simple example of using SVD with sklearn in Python. The following example uses the UCI ML iris dataset (since the tips dataset isn't ideal for SVD). The sklearn.decomposition.TruncatedSVD class allows to perform linear dimensionality reduction.

```python
# import libraries
from sklearn.decomposition import TruncatedSVD
from sklearn import datasets
import matplotlib.pyplot as plt

# load iris dataset as an example
iris = datasets.load_iris()
X = iris.data
y = iris.target

# SVD represent documents and terms in vectors 
svd = TruncatedSVD(n_components=2, algorithm='randomized', n_iter=100, random_state=122)
X_svd = svd.fit_transform(X)

# plot the result
plt.scatter(X_svd[:, 0], X_svd[:, 1], c=y, s=50, cmap='viridis')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.title('2-Component SVD')
plt.show()
```
