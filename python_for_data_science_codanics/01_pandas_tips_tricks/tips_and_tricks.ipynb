```python
df = pd.DataFrame({'A col':[1,2,3,6,7,8], 'B col':[4,5,6,32,2,9]})
df
```

```python
df.rename(columns={'A col':'col_a', 'B col': 'col_b'}, inplace=True)
df
```

```python
# another way
df.columns=['col_aa','col_bb']
df
```

```python
# to replace any character, string
#df.columns=['col_aa','col_bb'].str.replace('col','col_')
df.columns=df.columns.str.replace('col_','col-')
df
```

# 16 - Splitting a String into Multiple Columns

```python
# import libraries
import pandas as pd

df = pd.DataFrame({'name': ['Muhammad Saddam', 'Ahmad Fazal', 'Sajjad Khan', 'Abu Bakkar'],
                   'location': ['Lahore, Pakistan', 'Peshawar, Pakistan', 'Karachi, Pakistan', 'Islamabad, Pakistan']
                   })

df
```

```python
df.name.str.split(' ').head()
```

```python
# now expanding
df[['first_name','last_name']] = df.name.str.split(' ', expand=True)
df
```

```python
# now expanding the location

df.location.str.split(',',expand=True)
```

```python
df[['city','country']] = df.location.str.split(',',expand=True)
df
```

# 23 - Reshaping a Data Frame

```python
fasla = pd.DataFrame([['12345',100,200,300], ['34567',400,500,600], ['67890',700,800,900]],
                     columns=['zip','factory','warehouse','retail'])
fasla.head()
```

```python
fasla.head().T
```

```python
fasla_lang = fasla.melt(id_vars='zip', var_name='location_type', value_name='distance')
fasla_lang.head()
```

```python
import seaborn as sns

sns.barplot(x='zip', y='distance', hue='location_type', data=fasla_lang)
```
